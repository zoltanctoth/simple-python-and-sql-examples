```python
# Exercise
# 0. Notebook
# Create a new notebook and ensure it runs in the correct conda environment

# 1: Input
# Ask for a number (user keyboard input) and print it to the screen
# Hint: use the input() function 

# 2: Error handling 1
# Check what error is raised when the user inputs a non-number.
# Hint: You can use the int(...) function to convert the string result of input() to a number

# 3: Error handling 2
# Use a try-catch block to catch the error from (2.) and report on it with printing a simple error message

# 4: Random
# Generate a random number between 1 and 10 first, then add what you implemented in points (1.-2.-3.). If the random number is the same as the user input, say “You guessed it!”, otherwise say “You missed it”, and print the random number along with the user’s guess to the screen.

# 5: A Game
# Let’s create a guessing game. Python generates a random number between 1 and 10. The user guesses it (use input()) as long as it doesn’t hit it. If the user doesn’t guess it, respond if the original number is higher or lower than the guess. Make it error-proof with the error reporting you did (and add new error reporting if needed)
# Even if the input is invalid, print the user an error message and let them guess on!
# guessed = False
# while not guessed:
#     # make the user guess. If they guessed it then: guessed = True

# 6: Measure Completion
# At the end, print how many seconds it took the user to guess the number. Import the time.time() function under the name t_time() (as Claudia pointed out). Also print how many guesses it took to print the number.

# BONUS:
# Use your own NegativeNumberError error for negative numbers and your own NumberTooLargeError for catching inputs > 10. 

import random
from time import time as t_time


# 5. Game
class NegativeNumberError(Exception):
    pass

class NumberTooLargeError(Exception):
    pass

random_number = random.randint(1, 10)
start_time = t_time()
guesses = 0
guessed = False
while not guessed:
    try:
        user_input = int(input("Guess a number between 1 and 10: "))
        if user_input < 1:
            raise NegativeNumberError("Negative numbers are not allowed.")
        elif user_input > 10:
            raise NumberTooLargeError("Numbers greater than 10 are not allowed.")
    except ValueError:
        print("Invalid input. Please enter a number.")
    except NegativeNumberError as nne:
        print(nne)
    except NumberTooLargeError as ntle:
        print(ntle)

    guesses += 1
    if user_input == random_number:
        guessed = True
        print(f"You guessed it! The number was {random_number}.")
        # Next line: .2f means 2 decimal places
        print(f"It took you {t_time() - start_time:.2f} seconds to guess the number.")
        print(f"You made {guesses} guesses.")
    elif user_input < random_number:
        print("The number is higher.")
    else:
        print("The number is lower.")

    
```
